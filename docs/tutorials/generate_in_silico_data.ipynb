```python
from yeastdnnexplorer.probability_models.generate_data import (generate_gene_population, 
															   generate_perturbation_binding_data)
```

```python
# generate a population of 1000 genes with 30% of them labelled 1 (signal)
# and 0 (noise)
# note that 1000 and 0.3 are the default values for this function
population1 = generate_gene_population(1000, 0.3)

population1
```

```python
# generate perturbation and binding data based on this population
# note that these are the default values for this function
population1_tf1_data = generate_perturbation_binding_data(
	population1,
	0.0, 1.0,
	3.0, 1.0,
	1e-3,
	0.5)

# label this with an arbitrary TF name
population1_tf1_data['regulator'] = 'TF1'

population1_tf1_data
```
